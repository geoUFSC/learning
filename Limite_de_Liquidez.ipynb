```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the logarithmic function to fit
def log_function(N, a, b):
    return a * np.log(N) + b

# Define the data
N = np.array([13, 20, 29, 36])
w = np.array([0.268, 0.264, 0.258, 0.254])

# Perform the curve fitting
params, covariance = curve_fit(log_function, N, w)

# Extract the fitted parameters
a, b = params

# Create a range of N values for the fitted curve
N_fit = np.linspace(min(N), max(N), 100)

# Calculate the fitted values for the range of N
w_fit = log_function(N_fit, a, b)

# Calculate the fitted value for N=25
N_ll = 25
w_ll = log_function(N_ll, a, b)

# Plot the original data and the fitted curve
plt.scatter(N, w, label="Data")
plt.semilogx(N_fit, w_fit, 'r', label=f"Fitted Curve: {a:.4f}*log(N) + {b:.4f}")
plt.scatter(N_ll, w_ll, label=f"LL: {w_ll:.4f}")
plt.xlabel("N")
plt.ylabel("w")
plt.legend()
plt.grid()
plt.show()
```
